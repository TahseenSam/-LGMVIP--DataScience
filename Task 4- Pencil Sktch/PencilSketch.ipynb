# Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

# Read Your Image 

```python
img=cv2.imread('White-Daisies-600x402.jpg')
#type(img)
cv2.imshow("Daisy",img)
cv2.waitKey(0)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

### Convert BGR to Gray

```python
grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("GrayImg",grayImg)
cv2.waitKey(0)
```

### Convert  Gray to Image Negative

```python
negative_image = 255 - grayImg
cv2.imshow("Negative Image", negative_image)
cv2.waitKey(0)
```

### Apply Gaussian Blur

```python
blurred = cv2.GaussianBlur(negative_image, (25, 25), 0)
cv2.imshow("blurred", blurred)
cv2.waitKey(0)
```

### Converting Blurred Image To Sketch

```python
negative_blurred = 255 - blurred
pencil_sketch = cv2.divide(grayImg, negative_blurred, scale=256.0)
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```

```python
cv2.imshow("original image", img)
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```
